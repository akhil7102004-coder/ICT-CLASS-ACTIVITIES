```python
arr = np.random.rand(5,4)
arr
```

```python
row_sum = arr.sum(axis=1)

print("Row with max sum:", arr[np.argmax(row_sum)])
print()
row_min = arr.min(axis=1, keepdims=True)
row_max = arr.max(axis=1, keepdims=True)

normalized = (arr - row_min) / (row_max - row_min)

print("Normalized:\n\n",normalized)
```

```python
arr = np.arange(1,37).reshape(6,6)
sub = arr[1:5, 1:5]
even = sub[sub % 2 == 0]
print("Array:\n",arr)
print()
print("SubMatrix:\n",sub)
print()
print("All even numbers from this submatrix:\n",even)
```

```python
arr = np.arange(12).reshape(3,4)
print("Array:/\n",arr)
print()
res = arr.reshape(2,6)
print("Reshaped:\n",res)
print()
flat = res.flatten()
print("Flatten:\n",flat)

back = flat.reshape(4,3)
print()
print(" array back to shape:\n",back)
```
